```python
#ითვლის მსხვეპლთა რიცხვს ორივე ქვეყანაში წლების მიხედვით
year = df['date_of_death']
year.str.slice(0, 4).value_counts()
```

```python
#ქვეყნების მიხედვით მსხვერპლთა საშუალო ასაკი
group_citiz['age'].mean()
```

```python
#ქვეყნების მიხედვით მსხვერპლთა ასაკის მედიანა 
group_citiz['age'].median()
```

```python
import matplotlib.pyplot as plt

labels = 'Palestinian Man', 'Palestinian women', 'Israeli Man', 'Israeli Women'
I_W = int(df.loc[gender_F & israeli_citizen, ['name']].count())
I_M = int(df.loc[gender_M & israeli_citizen, ['name']].count())
P_W = int(df.loc[gender_F & palestinian_citizen, ['name']].count())
P_M = int(df.loc[gender_M & palestinian_citizen, ['name']].count())
sizes = [P_M, P_W, I_M, I_W]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

arr_injur = df['type_of_injury'].value_counts().index.tolist() 
arr_count = df['type_of_injury'].value_counts().to_numpy()

ax.bar(arr_injur, arr_count)
plt.xticks(rotation='vertical')
ax.set_ylabel('number of death')

plt.show()
```

```python
import matplotlib.pyplot as plt

year = df['date_of_death']
arr_year = year.str.slice(0, 4).value_counts().index.tolist() 
death_count = year.str.slice(0, 4).value_counts().to_numpy()
death_count

labels = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

plt.plot(arr_year, death_count)
plt.xticks(arr_year, labels, rotation='vertical')
plt.margins(0.2)
plt.subplots_adjust(bottom=0.15)
plt.show()
```
